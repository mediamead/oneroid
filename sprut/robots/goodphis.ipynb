```python
import pybullet as p
import numpy as np

from pybullet_robot import PyBulletRobot, W, H

import sys, glob
import pandas as pd 

import matplotlib.pyplot as plt
```

```python
gui = True
r = PyBulletRobot(4, 4, render=gui)
```

```python
# load csv
fnames = "goodphis-*.csv"
good_phiss = None
for fname in glob.glob(fnames):
    df = pd.read_csv(fname)
    if good_phiss is None:
        good_phiss = df
    else:
        good_phiss = pd.concat([good_phiss, df])
print(good_phiss.values.shape)

# take only S-shaped poses
_good_phiss = []
for i in range(1, good_phiss.shape[0]):
    phis = good_phiss.values[i,1:]
    phis = phis.reshape(-1, 2)

    r.step(phis)

    if phis[0][0] < 0:
        _good_phiss.append(phis)
good_phiss = np.array(_good_phiss)
```

```python
# show histogram of aggregate (over all joints) of X and Y angles
print(good_phiss.shape)
_ = plt.hist(good_phiss[:,:,0].sum(axis=1))
```

```python
print(good_phiss.shape)
_ = plt.hist(good_phiss[:,:,1].sum(axis=1))
```

```python
# extract cam_ps
cam_ps = np.zeros((len(good_phiss), 4))

for i in range(len(good_phiss)):
    r.step(good_phiss[i])
    cam_p, cam_v, cam_u = r.getHeadcamPVU()

    cam_ps[i,0:] = i
    cam_ps[i,1:] = cam_p
```

```python
# scapy plot of X & Y
_ = plt.plot(cam_ps[:,1], cam_ps[:,2], '.')
```

```python
# histograms of X, Y, and Z
_ = plt.hist(cam_ps[:,1], bins=50)
```

```python
_ = plt.hist(cam_ps[:,2], bins=50)
```

```python
_ = plt.hist(cam_ps[:,3], bins=50)
```

```python
# choose the most popular spot +- 5cm
cam_ps0 = []

for cam_p in cam_ps:
    if cam_p[1] > 0.3 and cam_p[1] < 0.4:
        if cam_p[2] > -0.05 and cam_p[2] < 0.05:
            if cam_p[3] > 0.65 and cam_p[3] < 0.7:
                cam_ps0.append(cam_p)
cam_ps0 = np.array(cam_ps0)
print(cam_ps0.shape)
```

```python
_ = plt.plot(cam_ps0[:,1], cam_ps0[:,2], '.')
```

```python
_ = plt.hist(cam_ps0[:,3], bins=50)
```

```python
import time
for cam_p0 in cam_ps0:
    i = int(cam_p0[0])
    phis = good_phiss[i]
    phis = phis.reshape(-1, 2)

    r.step(phis)
    r.getCameraImage()
    time.sleep(1)
```
