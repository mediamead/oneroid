```python
import numpy as np

a = 0.024
b = 0.035
r = np.sqrt(a*a + b*b)
beta = np.arctan(b/a)

def x(theta):
    """
    Given desired joint rotation angle, returns necessary string length
    """
    alpha = (np.pi - 2 * beta - theta) / 2
    #print("#theta=%f" % (theta / np.pi * 180))
    #print("#beta=%f" % (beta / np.pi * 180))
    #print("#alpha=%f" % (alpha / np.pi * 180))
    #print("#sum-pi=%f" % (2*beta + 2*alpha + theta - np.pi))
    a = 2 * r * np.sin(alpha)
    return a
#print("#theta=0, x-a=%f" % (x(0) - 2*a))    
```

```python
import matplotlib.pyplot as plt

thetas = np.arange(0, np.pi/4, np.pi/400)

def approx_x(theta):
    """
    Given desired joint rotation angle, returns necessary string length (linear approximation)
    """    
    k = (x(np.pi/4)-x(0)) / (np.pi/4)
    return k*theta + x(0)

_fig, ax = plt.subplots()
ax.plot(thetas, x(thetas), 'r', label='calculated')
ax.plot(thetas, approx_x(thetas), 'b', label='approximated')
legend = ax.legend(loc='upper right', shadow=True, fontsize='x-large')
plt.xlabel('theta')
plt.ylabel('x')
plt.show()
```

```python
plt.plot(thetas, (x(thetas) - approx_x(thetas)) / x(thetas)*100)
plt.xlabel('theta')
plt.ylabel('error %')
plt.show()
```
