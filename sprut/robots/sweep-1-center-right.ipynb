```python
zs = np.array(res[0])
pxyzs = np.array(res[1])
phiss = np.array(res[2])
costs = np.array(res[3])

import matplotlib.pyplot as plt
```

```python
x = pxyzs.squeeze(3)[:,1]
y = pxyzs.squeeze(3)[:,2]
z = pxyzs.squeeze(3)[:,3]
```

```python
def plot1(t):
    """
    all betas, all gammas values, thetas
    x, y, z, vectors
    """
    plt.subplot(231)
    plt.plot(t,phiss[:,0,0],'red', t,phiss[:,1,0],'yellow', t,phiss[:,2,0],'green',t,phiss[:,3,0],'blue')
    plt.ylabel('betas')
    plt.ylim(-0.5,1.)

    plt.subplot(232)
    plt.plot(t,phiss[:,0,1],'red', t,phiss[:,1,1],'yellow', t,phiss[:,2,1],'green',t,phiss[:,3,1],'blue')
    plt.ylabel('gamma')
    plt.ylim(-0.5,1.)

    plt.subplot(233)
    theta1 = np.arctan2(phiss[:,0,0], phiss[:,0,1])
    theta2 = np.arctan2(phiss[:,1,0], phiss[:,1,1])
    theta3 = np.arctan2(phiss[:,2,0], phiss[:,2,1])
    theta4 = np.arctan2(phiss[:,3,0], phiss[:,3,1])
    plt.plot(t,theta1,'red', t,theta2,'yellow', t,theta3,'green', t,theta4,'blue')
    plt.ylabel('thetas')

    plt.subplot(234)
    plt.plot(t,x[:,0],'red', t,x[:,1],'green', t,x[:,2],'blue')
    plt.ylabel('x')
    plt.ylim(-1.1,1.1)

    plt.subplot(235)
    plt.plot(t,y[:,0],'red', t,y[:,1],'green', t,y[:,2],'blue')
    plt.ylabel('y')
    plt.ylim(-1.1,1.1)

    plt.subplot(236)
    plt.plot(t,z[:,0],'red', t,z[:,1],'green', t,z[:,2],'blue')
    plt.ylabel('z')
    plt.ylim(-1.1,1.1)
```

```python
def plot2(t):
    plt.subplot(231)
    plt.plot(t,phiss[:,0,0]+phiss[:,1,0],'orange', t,phiss[:,2,0]+phiss[:,3,0],'cyan')
    plt.ylabel('betas sum')

    plt.subplot(232)
    plt.plot(t,phiss[:,0,1]+phiss[:,1,1],'orange', t,phiss[:,2,1]+phiss[:,3,1],'cyan')
    plt.ylabel('gamma sum')
    #plt.ylim(-0.5,1.)

    plt.subplot(233)
    plt.plot(t,phiss[:,0,1]+phiss[:,2,1],'orange', t,phiss[:,1,1]+phiss[:,3,1],'cyan')
    plt.ylabel('gamma sum2')
    #plt.ylim(-0.5,1.)

    plt.subplot(234)
    plt.plot(t,phiss[:,0,0]-phiss[:,1,0],'orange', t,phiss[:,2,0]-phiss[:,3,0],'cyan')
    plt.ylabel('betas sub')
    #plt.ylim(-1.,1.)

    plt.subplot(235)
    plt.plot(t,phiss[:,0,1]-phiss[:,1,1],'orange', t,phiss[:,2,1]-phiss[:,3,1],'cyan')
    plt.ylabel('gamma sub')
    #plt.ylim(-1.,1.)

    plt.subplot(236)
    plt.plot(t,phiss[:,0,1]-phiss[:,2,1],'orange', t,phiss[:,1,1]-phiss[:,3,1],'cyan')
    plt.ylabel('gamma sub2')
    #plt.ylim(-1.,1.)
```

```python
def plot3(t):
    """
    all betas, all gammas
    all thetas, all amps
    """
    plt.subplot(221)
    plt.plot(t,phiss[:,0,0],'red', t,phiss[:,1,0],'yellow', t,phiss[:,2,0],'green',t,phiss[:,3,0],'blue')
    plt.ylabel('betas')
    plt.ylim(-0.5,1.)

    plt.subplot(222)
    plt.plot(t,phiss[:,0,1],'red', t,phiss[:,1,1],'yellow', t,phiss[:,2,1],'green',t,phiss[:,3,1],'blue')
    plt.ylabel('gamma')
    plt.ylim(-0.5,1.)

    theta1 = np.arctan2(phiss[:,0,0], phiss[:,0,1])
    theta2 = np.arctan2(phiss[:,1,0], phiss[:,1,1])
    theta3 = np.arctan2(phiss[:,2,0], phiss[:,2,1])
    theta4 = np.arctan2(phiss[:,3,0], phiss[:,3,1])
    a1 = np.linalg.norm(phiss[:,0,:],axis=1)
    a2 = np.linalg.norm(phiss[:,1,:],axis=1)
    a3 = np.linalg.norm(phiss[:,2,:],axis=1)
    a4 = np.linalg.norm(phiss[:,3,:],axis=1)
    plt.subplot(223)
    plt.plot(t,theta1,'red', t,theta2,'yellow', t,theta3,'green', t,theta4,'blue')
    plt.ylabel('thetas')

    plt.subplot(224)
    plt.plot(t,a1,'red', t,a2,'yellow', t,a3,'green', t,a4,'blue')
    plt.ylabel('amps')
```

```python
plot3(t)
```
